## Nudity Detection Class

```python
nude_image = 'bikini.jpg'
censored_image = 'censored.jpg'
```

```python
from nudenet import NudeDetector
detector = NudeDetector()
```

```python
# Performing detection
print(detector.detect(nude_image))
```

```python
# Censoring an image
detector.censor(nude_image, out_path=censored_image, visualize=False)
```
